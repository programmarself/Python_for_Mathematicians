#Mean, Mode, Median and Standard Deviation

Mean, median, and mode are measures of central tendency used to summarize data, with the mean being the average, the median the middle value, and the mode the most frequent value. Standard deviation and variance measure data spread; variance quantifies the average squared deviation from the mean, while standard deviation (its square root) provides the average deviation in the same units as the data. These metrics are essential for understanding the data's central point, variability, and distribution, guiding insights, comparisons, and decision-making while addressing outliers and patterns effectively.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode

# Define a simple dataset
data = [2, 4, 4, 6, 8, 8, 8, 10]

# Calculate statistics
mean_value = np.mean(data)                  # Mean
median_value = np.median(data)              # Median
mode_result = mode(data)                    # Mode (returns a ModeResult object)
mode_value = mode_result.mode if mode_result.count > 1 else mode_result.mode[()]  # Extract mode correctly
std_dev_value = np.std(data, ddof=0)        # Standard Deviation (population)
variance_value = np.var(data, ddof=1)       # Variance (population)


# Visualization
plt.figure(figsize=(10, 6))

# Plot the data points as a scatter plot
plt.scatter(range(1, len(data) + 1), data, color='blue', label='Data Points', s=[30,40,50,60,70,80,90,100])                  #size= 100

# Add horizontal lines for statistics
plt.axvline(x=mean_value, color='red', linestyle='--', linewidth=2, label=f"Mean = {mean_value:.2f}")
plt.axvline(x=median_value, color='green', linestyle='-', linewidth=2, label=f"Median = {median_value}")
plt.axvline(x=mode_value, color='purple', linestyle='-.', linewidth=2, label=f"Mode = {mode_value}")
plt.axvline(x=std_dev_value, color='orange', linestyle=':', linewidth=2, label=f"Std Dev = {std_dev_value:.2f}")
plt.axvline(x=variance_value, color='brown', linestyle='-', linewidth=2, label=f"Variance = {variance_value:.2f}")

# Add labels, title, and legend
plt.title("Visualization of Mean, Median, Mode, and Standard Deviation", fontsize=12)
plt.xlabel("Index (i)", fontsize=9)
plt.ylabel("Value", fontsize=9)
plt.legend(fontsize=9)

# Show the plot
plt.tight_layout()
plt.show()
```

In regression analysis, the intercept represents the predicted value of the dependent variable when all independent variables are zero, providing a baseline for the relationship modeled by the equation. Intercepts are crucial in understanding and interpreting equations, models, and trends in various contexts.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example dataset (you can change this to any dataset)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # Independent variable (e.g., years, distance, etc.)
y = np.array([2, 4, 5, 4, 5])  # Dependent variable (e.g., price, temperature, etc.)

# Fit a linear regression model
model = LinearRegression()
model.fit(x, y)

# Extract the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Generate the fitted line
x_line = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_line = model.predict(x_line)

# Visualization
plt.figure(figsize=(8, 6))

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points', s=100)    #sizes = 100

# Plot the fitted line
plt.plot(x_line, y_line, color='red', label=f"Fitted Line: y = {slope:.2f}x + {intercept:.2f}")

# Highlight the intercept on the y-axis
plt.axhline(y=intercept, color='green', linestyle='--', label=f"Intercept = {intercept:.2f}")

# Add labels, title, and legend
plt.title("Linear Regression with Intercept Visualization", fontsize=16)
plt.xlabel("Independent Variable (x)", fontsize=12)
plt.ylabel("Dependent Variable (y)", fontsize=12)
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

This formula extends to higher dimensions by **squared root** of summing the **squared differences** of corresponding coordinates. Euclidean distance is widely used in geometry and as a similarity measure in machine learning. It provides an intuitive and simple metric for quantifying the spatial separation between points, making it fundamental in various analytical and computational applications.

```python
import numpy as np
import matplotlib.pyplot as plt

# Euclidean Distance function definition
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point2 - point1)**2))

# Example points for calculating Euclidean distance
point1 = np.array([1, 2])
point2 = np.array([4, 6])

# Calculate Euclidean distance
distance = euclidean_distance(point1, point2)

# Plotting the points and distance
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'ro-', label=f"Distance = {distance:.2f}")
plt.scatter(point1[0], point1[1], color='green', s=100, label="Point 1")
plt.scatter(point2[0], point2[1], color='orange', s=100, label="Point 2")
plt.title(f"Euclidean Distance = {distance:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()

# Print the calculated distance
print(f"Euclidean Distance: {distance:.2f}")
```
