#**Functions:**
Functions allow us to create blocks of code that can be easily executed many times without needing to constantly rewrite the entire block of code.

```python
def happy_birthday_song():
    print("happy birthday to you")
    print("happy birthday to you")
    print("happy birthday dear")
    print("happy birthday to you")
    print()

happy_birthday_song()


```

**Which One is Better?**

Using a **function** is better because it makes the code cleaner, more modular, and reusable. If you ever need to change the song’s wording or format, you can simply edit the function, and the changes will apply wherever it’s called.

Using a **loop** alone is less readable and harder to maintain, especially if the repeated code grows more complex.

# **What if we want to add name and age of birthday boy/girl?**

```python
def happy_birthday_song(name, age):
    print("happy birthday to you")
    print("happy birthday to you")
    print(f"happy birthday dear {name}")
    print(f"You are now {age} years old")
    print()

happy_birthday_song("Shafaq", 21)
happy_birthday_song("Fakiha", 23)
happy_birthday_song("Almas", 24)
```

# **The Gift of Return:**

A **return** in a function is like a "result" that the function gives back to the rest of your program. When a function is done running, it can send a value back to wherever it was called. This returned value can then be used or stored for later.

*   Think of it like ordering food at a restaurant: you give your order (call the function), the chef (the function) prepares it, and then they return your food (the result) for you to enjoy.

```python
def add(a, b):
  c = a + b
  return c

def subtract(a, b):
  c = a - b
  return c

def multiply(a, b):
  c = a * b
  return c

def divide(a, b):
  c = a / b
  return c

add(1,4)
```

```python
def create_name(first_name, last_name):
    full_name = first_name + " " + last_name
    return full_name.title()

musician = create_name('jimi', 'hendrix')
print(musician)

print(create_name('john', 'hooker'))
```

# **Temperature Converter**

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

# Get user input
temperature = float(input("Enter the temperature: "))
unit = input("Is this temperature in Celsius or Fahrenheit? (C/F): ").upper()

if unit == 'C':
    fahrenheit = celsius_to_fahrenheit(temperature)
    print(f"{temperature}°C is {fahrenheit:.2f}°F")
elif unit == 'F':
    celsius = fahrenheit_to_celsius(temperature)
    print(f"{temperature}°F is {celsius:.2f}°C")
else:
    print("Invalid unit. Please enter 'C' for Celsius or 'F' for Fahrenheit.")
```
