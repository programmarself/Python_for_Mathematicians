```python
import matplotlib.pyplot as plt
```

```python
plt.plot (5,5, marker='x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('My Plot')
plt.show()
```

```python
#example
x= [1,2,3,4,5]
y= [1,5,7,9,3]
plt.plot(x,y, marker = 'o', color= 'blue', label = 'y=2x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('My plot')
plt.grid(True)
plt.show()
```

```python
#example 2
x = ['Ali','Malik','Humza']
y = [2,4,6]
plt.bar(x,y, color= "orange")
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Students Marks')
plt.show()
```

```python
#example 3
#Scatter plt

x = [ 1,4,6,7,8]
y = [2,4,5,6,7]

plt.scatter(x,y, color = "green", marker = "x", label ="Scatter Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.plot(x,y, color = "green", marker = "x", label ="line", linestyle = '--')
plt.text(3,6, 'x',fontsize= 11, color = 'blue')
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np


# Creating dataset
np.random.seed(10)
data_1 = np.random.normal(100, 20, 200)
data_2 = np.random.normal(40,300,500)
data = [data_1, data_2]
fig = plt.figure(figsize =(10, 7))
#ax = fig.add_axes([0,1])
#bp = ax.boxplot(data)
# Creating plot
plt.boxplot(data)


# show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,5,6,7]
ax= plt.subplot()
sns.scatterplot(x=x,y=y, color = "blue", ax=ax)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.title('My plot')
plt.show()
```
